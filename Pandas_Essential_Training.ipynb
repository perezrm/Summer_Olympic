### Which country has won the omst men's gold in singles badminton over the years?
### Sort the results alphabetically by the player's names.

```python
# check to see if badminton is in the Sport colmnun
oo.Sport.value_counts()
```

```python
# multiple conditions encapsulate in square brackets

oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Sport == 'Badminton')]
```

```python
# Sort the results alphabetically by the player's names

gbm = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Sport == 'Badminton')]
gbm.sort_values(by = 'Athlete')
```

### Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
# conditional operator >= greater than or equal the other
# tail end of the data is 2008 no need to address

rec = oo[oo.Edition >= 1984]
rec.NOC.value_counts().head(3) # the top three countries .head()
```

### Display the male gold medal winners for the 100m Track & Field sprint event over the years.
### List the results starting with the most recent.
### Show the Olympic City, Edition, Athlete and the country they represent.

```python
mgh = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Event == '100m')]
mgh
```

```python
# Listed by most recent

mgh = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Event == '100m')]
mgh.sort_values('Edition', ascending=False)
```

```python
# Listed by most recent and by specific the columns

mgh = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Event == '100m')]
mgh.sort_values('Edition', ascending=False)[['City', 'Edition', 'Athlete', 'NOC']]
```
