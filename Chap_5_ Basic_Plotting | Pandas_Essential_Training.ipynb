## Colormaps | Sequential, Diverging, Qualitative
### The sequential should be used for representing information that has ordering. There is a change in lightness, often over a single hue.
### Deverging is to be used when the information being plotted deviates around a middle value. Here there are often two different colors being used.
### The qualitative class is used to represent information which does not have any ordering or relationship, and is oftern miscellaneous colors.

# ch. 9 Data Visualization 

### Using the Olympic dataset, present a summary of the total medals won by participating countries in the 2008 Olympics.

```python
lo = oo[oo.Edition == 2008]
lo
```

```python
g = lo.groupby(['NOC', 'Medal']).size().unstack('Medal', fill_value=0)
g = g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']] # when it's a list []
g # CALL IT
```

```python
g = lo.groupby(['NOC', 'Medal']).size().unstack('Medal', fill_value=0)
g = g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head().plot(kind='bar') # when it's a list []
g # CALL IT
```

```python
sns.heatmap(g)
```

```python
g = g.transpose()
g
```
